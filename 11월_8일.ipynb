문제 3 다음의 X,y를 이용하여 X_train, y_train, X_test, y_test를 나누기

```python
X = wine.data
y =wine.target
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state=42, stratify =y)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# train 데이터와 test 데이터 나눈후 아래 실행하여 결과 확인하기

print(X_test.shape)
print(y_test.shape)
```

k=3일때 KNN 이용--> test데이터에 대한 정확도 구하기

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics  # 성능지표
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train) #학습데이터로 학습하여 모델 최적화
y_predict = knn.predict(X_test) #최적화된 모델을 이용하여 테스트 데이터를 예측한다.
print('Accuracy: ', metrics.accuracy_score(y_test, y_predict) ) #y_test 실제값, y_predicted 모델을 통한 예측값
```

문제 5 for loop 사용하여 k에 따른 mean_error와 accuracy구하기

```python
error = []

for i in range(1,50):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(X_train, y_train)
  pred_i = knn.predict(X_test)
  error.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,50),error,color='red',linestyle='dashed',marker='o',
         markerfacecolor='blue',markersize=10)
plt.title("Error rate K value")
plt.xlabel("K value")
plt.ylabel("mean error")
plt.show()
```

문제 6 최적 조건 k에서 test 데이터에 대한 정확도 구하기

```python
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, y_train) #학습데이터로 학습하여 모델 최적화
y_predict = knn.predict(X_test) #최적화된 모델을 이용하여 테스트 데이터를 예측한다.
print('Accuracy: ', metrics.accuracy_score(y_test, y_predict) ) #y_test 실제값, y_predicted 모델을 통한 예측값
```
