```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris= load_iris()
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df
```

```python
df['target'] = iris.target
df
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
```

```python
X1=df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
```

```python
X1.head(10)
```

```python
Y1=df[['target']]
```

```python
Y1.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
x_std = StandardScaler().fit_transform(X1)
```

```python
x_std
```

공분산 행렬 구하기

```python
import numpy as np
```

```python
cov_matrix = np.cov(x_std.T)
eigen_val, eigen_vec = np.linalg.eig(cov_matrix)

np.sort(eigen_val)[::-1]
print("val:","\n",eigen_val)
print("val:","\n",eigen_vec)
```

```python
# explaned variances : eigen_val의 비율

explained_variances = []
sums = np.sum(eigen_val)

for v in eigen_val:
  explained_variances.append(v/sums)
print(explained_variances)
```
